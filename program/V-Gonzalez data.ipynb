```python
%pylab inline
import h5py
from matplotlib.colors import LogNorm
f = h5py.File("../data/v_gonzalez/sfr.z6.hdf5", 'r')
def printname(name):
     print name
f.visit(printname)
```

```python
group = f['z6']

logM   = group['logM'].value
logSFR = group['logSFR'].value

group1 = f['Info']
IMF    = group1['IMF'].value
Model  = group1['Model'].value

print "IMF = ", IMF, "\nModel = ", Model
```

```python
BINS = 100

AUX = histogram2d(logM, logSFR, bins=BINS)
#AUX = histogram2d(logM, logSFR, range=([[10.3, 13.0], [-3,4.2]]), bins=BINS)

#histogram2d(x, y, bins=10, range=None, normed=False, weights=None)

SFR_mean = []
SFR_per  = []
for j in range(BINS):
    aux =0
    for i in range(BINS):
        aux += 1.0*AUX[0][i,j]
    #print aux, percentile(AUX[0][:,j],50)
    SFR_per.append(array([percentile(AUX[0][:,j],25), percentile(AUX[0][:,j],50), percentile(AUX[0][:,j],75)]) )
#percentile(a, q, axis=None, out=None, overwrite_input=False, interpolation='linear', keepdims=False)
    SFR_mean.append( aux/(1.0*BINS) )
SFR_mean=np.array(SFR_mean)
SFR_per=np.array(SFR_per)

MassRange = []
for i in range(AUX[1].shape[0]-1):
    MassRange.append( (AUX[1][i+1]+AUX[1][i])/2 )       
MassRange = np.array(MassRange)
```

```python
semilogy( MassRange, SFR_mean)
semilogy( MassRange, SFR_per[:,1])
semilogy( MassRange, SFR_per[:,0])
semilogy( MassRange, SFR_per[:,2])
```

```python
figure(figsize=(9,7))
plt.plot(MassRange,log10(SFR_mean),label='Mean Value')

plt.plot(MassRange,log10(SFR_per[:,1]), label='Percentile 50')
fill_between(MassRange,log10(SFR_per[:,0]),log10(SFR_per[:,2]), facecolor="red",alpha=0.3)
plt.yscale('linear', nonposy='clip')
plt.xlim([10.3,13])

legend(loc='upper left')

xlabel(r"$\log_{10}( M_h ) [M_\odot]$", fontsize=20)
ylabel(r"Something strange that is not $\log_{10}($SFR $)[M_\odot yr^{-1}]$", fontsize=20)
```

```python
data  = open( 'analysis/gonzalez_data.dat', 'w')

data.write('# Halo Mass SFR_mean SFR_per25 SFR_per50 SFR_per75\n')
for i in range(MassRange.size):
    if(MassRange[i]<=13):
        data.write(str(MassRange[i])+'\t'+str(SFR_mean[i])+'\t'+str(SFR_per[i,0])+'\t'+str(SFR_per[i,1])+'\t'+str(SFR_per[i,2])+'\n')
data.close()
```
