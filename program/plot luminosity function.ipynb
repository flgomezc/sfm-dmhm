# Star Formation Rate

Load the best parameters

```python
if( Dust_Ext ==1 ):
    Param_Scattering = np.loadtxt('analysis/best_parameters_w_ext.dat',usecols=(0,1,2,3), skiprows=0); 
else:
    Param_Scattering = np.loadtxt('analysis/best_parameters_wo_ext.dat',usecols=(0,1,2,3), skiprows=0); 

L_0 = Param_Scattering[:,0]
M_0 = Param_Scattering[:,1]
beta = Param_Scattering[:,2]
gamma = Param_Scattering[:,3]
```

Plotting the Star Formation Rate function using the best fit parameters

```python
#############################################
#                                           #
#           Plotting SFR function           #
#                                           #
#############################################
Mass = linspace(10.5,13,50)

figure(figsize(9,7))
for k in range(L_0.size):
    SFR = zeros(Mass.size)
    for i in range(Mass.size):
        SFR[i] = StarFormationRate(10**Mass[i],10**L_0[k],10**M_0[k],beta[k],gamma[k])
    loglog(10**Mass,SFR)
xlabel("DMH mass$[M_\odot]$", fontsize=20)
ylabel("SFR $[M_\odot yr^{-1}]$", fontsize=20)

#savefig("star_formation_rate.pdf")
```
