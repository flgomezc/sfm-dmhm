```python
A = range(0,64)
#output = [open( '../results/%i.dat' %filenumber, 'w') for filenumber in A]
Dust_Extinction()
if (Dust_Ext == 1): BestPar  = open( 'analysis/best_parameters_w_ext '+name+'.dat', 'w')
else:               BestPar  = open( 'analysis/best_parameters_wo_ext '+name+'.dat', 'w')

for DATA in A:
    if (Dust_Ext == 1): STR = 'results_w_ext '+name+'/'+str(DATA)+'.dat'
    else:               STR = 'results_wo_ext '+name+'/'+str(DATA)+'.dat'
        
    MCMC_raw = np.loadtxt(STR,usecols=(0,1,2,3,4,5), skiprows=0); 
    BoxLength = 250.0 # Mpc h^{-1}

    histo_0 = histogram(MCMC_raw[:,0], bins=30)
    histo_1 = histogram(MCMC_raw[:,1], bins=30)
    histo_2 = histogram(MCMC_raw[:,2], bins=30)
    histo_3 = histogram(MCMC_raw[:,3], bins=30)

    Lm = where( histo_0[0]==histo_0[0].max())[0][0]
    Mm = where( histo_1[0]==histo_1[0].max())[0][0]
    Bm = where( histo_2[0]==histo_2[0].max())[0][0]
    Gm = where( histo_3[0]==histo_3[0].max())[0][0]
    
    LM = (histo_0[1][Lm]+ histo_0[1][Lm+1])/2.0
    MM = (histo_1[1][Mm]+ histo_1[1][Mm+1])/2.0
    BM = (histo_2[1][Bm]+ histo_2[1][Bm+1])/2.0
    GM = (histo_3[1][Gm]+ histo_3[1][Gm+1])/2.0
    
    if (DATA==0):
        BestPar.write('# L_0 \t\tM_0 \t\tbeta \t\tgamma \t Box \n')
    BestPar.write(
               str(LM)    +'\t'+
               str(MM)    +'\t'+
               str(BM)    +'\t'+
               str(GM)    +'\t'+
               str(DATA)  +'\n')
    
BestPar.close()


if (Dust_Ext == 1):
    Param_Scattering = np.loadtxt('analysis/best_parameters_w_ext '+name+'.dat',usecols=(0,1,2,3), skiprows=0); 
else:
    Param_Scattering = np.loadtxt('analysis/best_parameters_wo_ext '+name+'.dat',usecols=(0,1,2,3), skiprows=0); 

L = Param_Scattering[:,0]
M = Param_Scattering[:,1]
B = Param_Scattering[:,2]
G = Param_Scattering[:,3]




fig = figure(figsize=(10, 9))
#fig.suptitle("Cosmic Variance", fontsize=30)
#text(.5,0.9,'Cosmic Variance', fontsize=30, horizontalalignment='center')

########################################################################## CHANGE FILENAME HERE !!!
fig.suptitle('Best parameters for '+name+' over 64 boxes.', fontsize=18)
#fig.subplots_adjust(hspace=0.20, wspace=0.20,top=1.5, right=1.0, bottom=0.0)
fig.subplots_adjust(hspace=0.25, wspace=0.20,top=0.91, right=0.95, bottom=0.08, left=0.07)

ax1 = fig.add_subplot(221)
ax1.hist(L)
ax1.set_xlabel("$\log_{10}(L_0)$",fontsize=20)

ax2 = fig.add_subplot(222)
ax2.hist(M)
ax2.set_xlabel("$\log_{10}(M_0)$",fontsize=20)
#ax2.yaxis.set_major_formatter(plt.NullFormatter())

ax3 = fig.add_subplot(223)
ax3.hist(B)
ax3.set_xlabel(r" $\beta $  ",fontsize=20)

ax4 = fig.add_subplot(224)
ax4.hist(G)
ax4.set_xlabel(r"$\gamma$",fontsize=20)
#ax4.yaxis.set_major_formatter(plt.NullFormatter())

########################################################################## CHANGE FILENAME HERE !!!
if (Dust_Ext == 1):
    savefig('analysis/bestpar_w '+name+'.png')
else:
    savefig('analysis/bestpar_wo '+name+'.png')
```
