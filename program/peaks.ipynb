```python
##################  Name of the Observed Dataset
name = "OD4"
```

```python
A = range(0,64)  #range(0,64)
#output = [open( '../results/%i.dat' %filenumber, 'w') for filenumber in A]

peaks10 = np.zeros([len(A),4])

Dust_Extinction()

#if (Dust_Ext == 1): BestPar  = open( 'analysis/best_parameters_w_ext '+name+'.dat', 'w')
#else:               BestPar  = open( 'analysis/best_parameters_wo_ext '+name+'.dat', 'w')

Param_list = ['L_0','M_0','alpha','gamma']
Param_labels = [r'$\log_{10}(L_0)$', r'$\log_{10}(M_0)$', r'$\alpha$', r'$\gamma$']


for DATA in A:
    if (Dust_Ext == 1): STR = 'results_w_ext '+name+'/'+str(DATA)+'.dat'
    else:               STR = 'results_wo_ext '+name+'/'+str(DATA)+'.dat'
        
    MCMC_raw = np.loadtxt(STR,usecols=(0,1,2,3,4,5), skiprows=0); 
    BoxLength = 250.0 # Mpc h^{-1}


    histo  = []
    Par_min = []
    index_min = where( MCMC_raw[:,4] == MCMC_raw[:,4].min() )[0] 
    for k in range(4):
        histo.append(histogram(MCMC_raw[:,k], bins=30))
        Par_min.append(MCMC_raw[index_min,k][0])
    
    
    fig = figure(figsize=(12, 9))

    fig.subplots_adjust(hspace=0.20, wspace=0.10,top=0.9, left=0.07, right=0.93, bottom=0.07)

    ax1 = fig.add_subplot(221)
    ax2 = fig.add_subplot(222)
    ax3 = fig.add_subplot(223)
    ax4 = fig.add_subplot(224)
        
    ax = [ax1, ax2, ax3, ax4]
    
    for k in range(4): 
####### Here we get the 0.10 covered region
        F_Lmax = histo[k][0].max()
        N = 1.0*MCMC_raw[:,0].size
        over = np.zeros(histo[k][0].size)
        F10=0
        F10_min = 0
        F10_max = 0
        
        for i in range(F_Lmax):
            for j in range(over.size):
                if histo[k][0][j]==F_Lmax-i :
                    over[j]=1

            covered_area = 0.0
            for j in range(over.size):
                covered_area += (histo[k][0][j]-F_Lmax+i)*over[j]/N

            if covered_area >= 0.10 :
                F63 = F_Lmax - i
                for j in range(over.size):
                    if over[j]==1:
                        F10_min = histo[k][1][j]
                        break
                for j in range(over.size):
                    if over[over.size-j-1]==1:
                        F10_max = histo[k][1][over.size-j]
                        break
                break
        
        if( (F10_min<Par_min[k])&(Par_min[k]<F10_max) ):
           peaks10[DATA,k]=1
#        print Param_list[k], F10_min, Par_min[k], F10_max, F10
        
########### Plotting        
        ax[k].hist(MCMC_raw[:,k],bins=30)
        ax[k].axvline(x=Par_min[k])
        ax[k].axvline(x=average(MCMC_raw[:,k]), color="red")
        ax[k].set_xlabel(str(Param_labels[k]),fontsize=20)   
        
        ax[k].axhline(y=F_Lmax-i, color="k")         
        ax[k].axvline(x=F10_min, linestyle='--')
        ax[k].axvline(x=F10_max, linestyle='--')
        
    if (Dust_Ext == 1):
        fig.suptitle("Parameter Dispersion for box #"+str(DATA)+"\n"+str(name)+" with Dust Attenuation", fontsize=24)
        savefig('peaks/'+name+'_w_'+str(DATA)+'.pdf')
    else:
        fig.suptitle("Parameter Dispersion for box #"+str(DATA)+"\n"+str(name)+" without Dust Attenuation", fontsize=24)    
        savefig('peaks/'+name+'_wo_'+str(DATA)+'.pdf')
############# End Plotting    
    

if (Dust_Ext == 1): file = open('peaks/peaks_'+name+'_w.dat', 'w')
else:               file = open('peaks/peaks_'+name+'_wo.dat', 'w')
file.write('# Log(L0), Log(M0), alpha, gamma\n')

for i in range(len(A)):
    for k in range(4):
        file.write(str(peaks10[i,k])+"\t")
    file.write("\n")
file.close()
    
```
