Willott data has equally spaced bins centered on

$-22.5,-22.0,-21.5,-21.0,-20.5$. 

In this study the bins must be defined as:

```python
HIST0_bins=[
-22.75,
    # -22.50
-22.25,
    # -22.00
-21.75,
    # -21.50
-21,25,
    # -21.00
-20,75,
    # -20.50
-20,25]
```

The x-error bar can be reported as the width of the bin: $\pm 0.25$

The data from Willott are (as reported in table 3):

```python
Willott = np.array( 
[# Mag,LumFunct, low_err, upp_err #
[-22.5, 2.66E-8, 9.08E-9, 7.78E-8],
[-22.0, 2.18E-6, 8.70E-7, 9.70E-6],
[-21.5, 1.45E-5, 2.88E-6, 2.92E-5],
[-21.0, 1.29E-4, 7.06E-5, 2.19E-4],
[-20.5, 2.30E-4, 9.34E-5,5.77E-4]])  

Obs_data1 = np.zeros([5,4])
Obs_data1_log10 = np.zeros([5,4])

Obs_data1yerror = [array(Willott[:,1]-Willott[:,2]),array(Willott[:,3]-Willott[:,1])]

for i in range(5):
    Obs_data1[i,0]=Willott[i,0]
    Obs_data1[i,1]=Willott[i,1]
    Obs_data1[i,2]=0
    Obs_data1[i,3]=0.25
    Obs_data1_log10[i,0]=Willott[i,0]
    Obs_data1_log10[i,1]=log10(Willott[i,1])
    Obs_data1_log10[i,2]=0
    Obs_data1_log10[i,3]=0.25
    
Obs_data1_log10yerror = [array(log10(Willott[:,1]/Willott[:,2])),array(log10(Willott[:,3]/Willott[:,1]))]
```

http://matplotlib.org/1.3.1/examples/statistics/errorbar_demo_features.html

Errorbars with Asymmetric error.

```python
figure(figsize(9,7))
plot()
ylabel(r"$\Phi(M_{1350})$ $($ Mpc$^{-3}$mag$^{-1})$",fontsize=20)
xlabel('$M_{1350}$',fontsize=20)
errorbar(Obs_data1_log10[:,0],Obs_data1_log10[:,1], xerr=0.25, yerr=Obs_data1_log10yerror, fmt='ro')
plt.grid(True)
```

```python
figure(figsize(9,7))
semilogy()
ylabel(r"$\Phi(M_{1350})$ $($ Mpc$^{-3}$mag$^{-1})$",fontsize=20)
xlabel('$M_{1350}$',fontsize=20)
errorbar(Obs_data1[:,0],Obs_data1[:,1], xerr=0.25, yerr=Obs_data1yerror, fmt='ko')
plt.grid(True)
```
