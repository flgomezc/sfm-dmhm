```python
# Last modification: 2014-apr-23
%pylab inline

from random import *
from observational_data import *
# Obs_data1: Willot  et al 2013
# obs_data2: Bouwens et al 2014
# obs_data3: McLure  et al 2009

# Open the raw data from Markov Chain-MoteCarlo
# L_0 	 M_0 	 beta 	 gamma 	 chi_sqr 	 Number of Bins
MCMC_raw = np.loadtxt('../results/0.dat',
                      usecols=(0,1,2,3,4,5), 
                      skiprows=0); 
BoxLength = 250.0 # Mpc h^{-1}

index = where( MCMC_raw[:,4] == MCMC_raw[:,4].min() )[0] 
L_min     = MCMC_raw[index,0][0]
M_min     = MCMC_raw[index,1][0]
beta_min  = MCMC_raw[index,2][0]
gamma_min = MCMC_raw[index,3][0]
```

```python
fig = figure(figsize=(10, 5))
fig.subplots_adjust(hspace=0.20, wspace=0.0,top=1.5, right=1.0, bottom=0.0)
ax1 = fig.add_subplot(221)
ax1.scatter(MCMC_raw[:,0], MCMC_raw[:,4], s=0.1)
ax1.set_xlabel("$\log_{10}(L_0)$",fontsize=20)
ax1.set_ylabel(" $\chi^2$ ",fontsize=20)
#ax1.set_ylim(0,5)
#ax1.axis((17.4,17.9,0,5))
#ax1.set_yscale=("log")


ax2 = fig.add_subplot(222)
ax2.scatter(MCMC_raw[:,1], MCMC_raw[:,4], s=0.1)
ax2.set_xlabel("$\log_{10}(M_0)$",fontsize=20)
#ax2.set_ylabel(" $\chi^2$ ",fontsize=20)
#ax2.set_ylim(0,5)
#ax2.axis((10.90,11.5,0,5))
#ax2.set_yscale=("log")
ax2.yaxis.set_major_formatter(plt.NullFormatter())

ax3 = fig.add_subplot(223)
ax3.scatter(MCMC_raw[:,2], MCMC_raw[:,4], s=0.1)
ax3.set_xlabel(r" $\beta $  ",fontsize=20)
ax3.set_ylabel(" $\chi^2$ ",fontsize=20)
#ax3.axis((0.0,2.0,0,5))
#ax3.set_ylim(0,5)
#ax3.set_yscale=("log")
#ax.yaxis.set_major_formatter(plt.NullFormatter())

ax4 = fig.add_subplot(224)
ax4.scatter(MCMC_raw[:,3], MCMC_raw[:,4], s=0.1)
ax4.set_xlabel(r"$\gamma$",fontsize=20)
#ax4.set_ylabel(" $\chi^2$ ",fontsize=20)
ax4.yaxis.set_major_formatter(plt.NullFormatter())
```

```python
output = open( '../analysis/0.dat', 'w' )
for i in range(MCMC_raw.shape[0]):
    if ( (MCMC_raw[i,0] <= L_min + 0.03)&(MCMC_raw[i,0] >= L_min - 0.05)):
        if ( (MCMC_raw[i,1] <= M_min + 0.05)&(MCMC_raw[i,1] >= M_min - 0.05)):
            if ( (MCMC_raw[i,2] <= beta_min + 0.1)&(MCMC_raw[i,2] >= beta_min - 0.1)):
                if ( (MCMC_raw[i,3] <= gamma_min + 0.1)&(MCMC_raw[i,3] >= gamma_min - 0.1)):
                    output.write(str(MCMC_raw[i,0])+"\t"+
                                 str(MCMC_raw[i,1])+"\t"+
                                 str(MCMC_raw[i,2])+"\t"+
                                 str(MCMC_raw[i,3])+"\t"+
                                 str(MCMC_raw[i,4])+"\t"+
                                 str(MCMC_raw[i,5])+"\n")
output.close()
```
