# Color Scale:

Related with the number of used bins during the fitting.

Red   = 21 bins

White = 15 bins

Blue  = 12 bins

```python
# Last modification: 2014-mar-30
%pylab inline
```

```python
data_mcmc = np.loadtxt('../results/60.dat', usecols=(0,1,2,3,4,5))
# L_0, M_0, beta, gamma, chi_sqr



V = data_mcmc[:,0] # L_0
W = data_mcmc[:,1] # M_0
X = data_mcmc[:,2] # beta
Y = data_mcmc[:,3] # gamma
Z = data_mcmc[:,4] # chi_sqr

index1 = where(data_mcmc[:,5]==21)
index2 = where(data_mcmc[:,5]==20)
index3 = where(data_mcmc[:,5]<20)

K = data_mcmc[:,5]

#K[index1]=1
#K[index2]=0.5
#K[index3]=0.0

colors = plt.cm.coolwarm(1.0*(K[:]-12)/9.0)


fig = figure(figsize=(10, 5))
fig.subplots_adjust(hspace=0.20, wspace=0.20,top=1.5, right=1.0, bottom=0.0)
ax1 = fig.add_subplot(221, axisbg='grey')
ax1.scatter(V, Z, s=0.2, edgecolors=colors)
ax1.set_xlabel("$\log_{10}(L_0)$",fontsize=20)
ax1.set_ylabel(" $\chi^2$ ",fontsize=20)
ax1.set_ylim(0,5)
#ax1.axis((17.4,17.9,0,5))
#ax1.set_yscale=("log")


ax2 = fig.add_subplot(222,axisbg='grey')
ax2.scatter(W, Z, s=0.2, edgecolors=colors)
ax2.set_xlabel("$\log_{10}(M_0)$",fontsize=20)
ax2.set_ylabel(" $\chi^2$ ",fontsize=20)
ax2.set_ylim(0,5)
#ax2.axis((10.90,11.5,0,5))
#ax2.set_yscale=("log")

#ax2.yaxis.set_major_formatter(plt.NullFormatter())

ax3 = fig.add_subplot(223,axisbg='grey')
ax3.scatter(X, Z, s=0.2, edgecolors=colors)
ax3.set_xlabel(r" $\beta $  ",fontsize=20)
ax3.set_ylabel(" $\chi^2$ ",fontsize=20)
#ax3.axis((0.0,2.0,0,5))
ax3.set_ylim(0,5)
#ax3.set_yscale=("log")
#ax.yaxis.set_major_formatter(plt.NullFormatter())

ax4 = fig.add_subplot(224,axisbg='grey')
ax4.scatter(Y, Z, s=0.2, edgecolors=colors)
ax4.set_xlabel(r"$\gamma$",fontsize=20)
ax4.set_ylabel(" $\chi^2$ ",fontsize=20)
ax4.set_ylim(0,5)
#ax4.axis((0.0,0.7,0.0,5.0))
#ax.set_yscale=("log")

#ax.yaxis.set_major_formatter(plt.NullFormatter())

```
