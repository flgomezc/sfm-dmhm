```python
%pylab inline
hpl  = 0.6777          # Using h from 2013-Planck

from random import *
from observational_data import *

def Luminosity(M,L_0,M_0,beta,gamma):
    return L_0*M*( (M/M_0)**(-beta) + (M/M_0)**gamma)**(-1)

def StarFormationRate(M,L_0,M_0,beta,gamma):
    return Luminosity(M,L_0,M_0,beta,gamma) * 1.4e-28
```

```python
#############################################
#            Loading data file              #
#############################################
# Data columns:
M = np.loadtxt('../data/MD_3840_Planck1/BDM/reduced_catshortV_007.dat'
    ,usecols=(3,),skiprows=17)/hpl
Mb = np.loadtxt('../data/MD_3840_Planck1/BDM/new_reduced_catshortV_007.dat'
    ,usecols=(3,),skiprows=17)/hpl
BoxLength = 1000.0/hpl # Mpc
```

```python
L       = np.zeros(M.size)
Lb      = np.zeros(Mb.size)
```

```python
################################################################
#                       Plotting the                           #
#                     fitting function                         #
################################################################

parameters_fb = loadtxt('MCMC_reg_fullbox.dat',usecols=(0,1,2,3,4,5), skiprows=0); 
index = where(parameters_fb[:,4]==parameters_fb[:,4].min() )
```

```python
L_0   = 10.0**parameters_fb[index,0]
M_0   = 10.0**parameters_fb[index,1]
beta  =  parameters_fb[index,2]
gamma =  parameters_fb[index,3]

#Calculate the histogram and Xi_sqr for first time.
#for i in range(M.size):
L[:] = Luminosity( M[:], L_0, M_0, beta, gamma)
Lb[:]= Luminosity(Mb[:], L_0, M_0, beta, gamma)
Magnitude_UV_galaxy_list = 51.82 - 2.5 * log10(L[:])
Magnitude_UV_galaxy_listb= 51.82 - 2.5 * log10(Lb[:])


Mag_Max=-22.65;Mag_Min=-16.0;Mag_bins=50.0;Mag_Stp= abs(Mag_Max-Mag_Min)/Mag_bins
A = linspace(Mag_Max,Mag_Min,Mag_bins);B = np.zeros(A.size-1)
for i in range(B.size):    B[i]=(A[i]+A[i+1])/2.0
HIST = 1.0*histogram(Magnitude_UV_galaxy_list,bins=A)[0]
for i in range(len(HIST)):
    HIST[i] = HIST[i]/( Mag_Stp*BoxLength**3)

HIST1 = 1.0*histogram(Magnitude_UV_galaxy_list,bins=HIST1_bins)[0]
HIST2 = 1.0*histogram(Magnitude_UV_galaxy_list,bins=HIST2_bins)[0]
HIST3 = 1.0*histogram(Magnitude_UV_galaxy_list,bins=HIST3_bins)[0]

HIST1b = 1.0*histogram(Magnitude_UV_galaxy_listb,bins=HIST1_bins)[0]
HIST2b = 1.0*histogram(Magnitude_UV_galaxy_listb,bins=HIST2_bins)[0]
HIST3b = 1.0*histogram(Magnitude_UV_galaxy_listb,bins=HIST3_bins)[0]

### Normalize the histograms
for i in range(len(HIST1)):
    HIST1[i] = HIST1[i]/((HIST1_bins[i+1]-HIST1_bins[i])*BoxLength**3)
    HIST1b[i]=HIST1b[i]/((HIST1_bins[i+1]-HIST1_bins[i])*BoxLength**3)
for i in range(len(HIST2)):
    HIST2[i] = HIST2[i]*1.0/(1.0*(HIST2_bins[i+1]-HIST2_bins[i])*BoxLength**3)
    HIST2b[i]=HIST2b[i]*1.0/(1.0*(HIST2_bins[i+1]-HIST2_bins[i])*BoxLength**3)
for i in range(len(HIST3)):
    HIST3[i] = HIST3[i]/((HIST3_bins[i+1]-HIST3_bins[i])*BoxLength**3)
    HIST3b[i]=HIST3b[i]/((HIST3_bins[i+1]-HIST3_bins[i])*BoxLength**3)
```
