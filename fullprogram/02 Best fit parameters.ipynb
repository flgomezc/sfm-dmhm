```python
%pylab inline

from random import *
from observational_data import *

def Luminosity(M,L_0,M_0,beta,gamma):
    return L_0*M*( (M/M_0)**(-beta) + (M/M_0)**gamma)**(-1)

def StarFormationRate(M,L_0,M_0,beta,gamma):
    return Luminosity(M,L_0,M_0,beta,gamma) * 1.4e-28
hpl  = 0.6777          # Using h from 2013-Planck
```

```python
A = range(0,64)
BestPar  = open( 'analysis/best_parameters.dat', 'w')

for DATA in A:

    STR = 'results/'+str(DATA)+'.dat'
    MCMC_raw = np.loadtxt(STR,usecols=(0,1,2,3,4,5), skiprows=0); 
    
    ChiSqr_min= MCMC_raw[:,4].min()
    index = where( MCMC_raw[:,4] == MCMC_raw[:,4].min() )[0] 
    L_min     = MCMC_raw[index,0][0]
    M_min     = MCMC_raw[index,1][0]
    beta_min  = MCMC_raw[index,2][0]
    gamma_min = MCMC_raw[index,3][0]
    ChiSqr_min= MCMC_raw[index,4][0]
    COLS_min  = MCMC_raw[index,5][0]
        
    BestPar.write( str(L_min)     +'\t'+
                   str(M_min)     +'\t'+
                   str(beta_min)  +'\t'+
                   str(gamma_min) +'\t'+
                   str(ChiSqr_min)+'\t'+
                   str(COLS_min)  +'\t'+
                   str(DATA)  +'\n')
BestPar.close()
```

```python
Param_Scattering = np.loadtxt('analysis/best_parameters.dat',usecols=(0,1,2,3,4,5), skiprows=0); 

chi_min = 0.5
chi_max = 2.0

Lum = Param_Scattering[:,0]
Mas = Param_Scattering[:,1]
B   = Param_Scattering[:,2]
G   = Param_Scattering[:,3]
C   = Param_Scattering[:,4]

#MCMC fullbox raw data
MCMCfb_raw = np.loadtxt("MCMC_reg_fullbox.dat",usecols=(0,1,2,3,4,5), skiprows=0);

ChiSqr_min= MCMC_raw[:,4].min()
index = where( MCMC_raw[:,4] == MCMC_raw[:,4].min() )[0] 
Lum_min     = MCMC_raw[index,0][0]
Mas_min     = MCMC_raw[index,1][0]
beta_min  = MCMC_raw[index,2][0]
gamma_min = MCMC_raw[index,3][0]
ChiSqr_min= MCMC_raw[index,4][0]
COLS_min  = MCMC_raw[index,5][0]

fig = figure(figsize=(10, 5))
#text(.5,0.9,'Cosmic Variance', fontsize=30, horizontalalignment='center')

fig.subplots_adjust(hspace=0.20, wspace=0.0,top=1.5, right=1.0, bottom=0.0)

ax1 = fig.add_subplot(221)
ax1.scatter(Lum,C, s=1.0)
ax1.set_xlabel("$\log_{10}(L_0)$",fontsize=20)
ax1.set_ylabel(" $\chi^2$ ",fontsize=20)
ax1.axvline(x=Lum_min)
ax1.axhline(y=ChiSqr_min)
ax1.axis([Lum.min(),Lum.max(),chi_min,chi_max])

ax2 = fig.add_subplot(222)
ax2.scatter(Mas, C, s=1.2)
ax2.set_xlabel("$\log_{10}(M_0)$",fontsize=20)
ax2.axvline(x=Mas_min)
ax2.yaxis.set_major_formatter(plt.NullFormatter())
ax2.axhline(y=ChiSqr_min)
ax2.axis([Mas.min(),Mas.max(),chi_min,chi_max])


ax3 = fig.add_subplot(223)
ax3.scatter(B, C, s=1.0)
ax3.set_xlabel(r" $\beta $  ",fontsize=20)
ax3.set_ylabel(" $\chi^2$ ",fontsize=20)
ax3.axvline(x=beta_min)
ax3.axhline(y=ChiSqr_min)
ax3.axis([B.min(),B.max(),chi_min,chi_max])

ax4 = fig.add_subplot(224)
ax4.scatter(G, C, s=1.0)
ax4.set_xlabel(r"$\gamma$",fontsize=20)
ax4.axvline(x=gamma_min)
ax4.axhline(y=ChiSqr_min)
ax4.yaxis.set_major_formatter(plt.NullFormatter())
ax4.axis([G.min(),G.max(),chi_min,chi_max])
```
