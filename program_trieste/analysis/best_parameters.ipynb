```python
# Open the raw data from Markov Chain-MoteCarlo
# L_0 	 M_0 	 beta 	 gamma 	 chi_sqr 	 Number of Bins
MCMC_raw = np.loadtxt('../results/6.dat',
                      usecols=(0,1,2,3,4,5), 
                      skiprows=0); 
BoxLength = 250.0 # Mpc h^{-1}

index = where( MCMC_raw[:,4] == MCMC_raw[:,4].min() )[0] 
L_min     = MCMC_raw[index,0][0]
M_min     = MCMC_raw[index,1][0]
beta_min  = MCMC_raw[index,2][0]
gamma_min = MCMC_raw[index,3][0]


histo_0 = histogram(MCMC_raw[:,0], bins=30)
histo_1 = histogram(MCMC_raw[:,1], bins=30)
histo_2 = histogram(MCMC_raw[:,2], bins=30)
histo_3 = histogram(MCMC_raw[:,3], bins=30)

Lm = where( histo_0[0]==histo_0[0].max())[0][0]
Mm = where( histo_1[0]==histo_1[0].max())[0][0]
Bm = where( histo_2[0]==histo_2[0].max())[0][0]
Gm = where( histo_3[0]==histo_3[0].max())[0][0]

LM = (histo_0[1][Lm]+ histo_0[1][Lm+1])/2.0
MM = (histo_1[1][Mm]+ histo_1[1][Mm+1])/2.0
BM = (histo_2[1][Bm]+ histo_2[1][Bm+1])/2.0
GM = (histo_3[1][Gm]+ histo_3[1][Gm+1])/2.0

print LM, MM, BM, GM
```

```python


############## Scattering plots ########################

fig = figure(figsize=(10, 5))
fig.subplots_adjust(hspace=0.20, wspace=0.0,top=1.5, right=1.0, bottom=0.0)
ax1 = fig.add_subplot(221)
ax1.scatter(MCMC_raw[:,0], MCMC_raw[:,4], s=0.1)
ax1.set_xlabel("$\log_{10}(L_0)$",fontsize=20)
ax1.set_ylabel(" $\chi^2$ ",fontsize=20)
#ax1.set_ylim(0,5)
#ax1.axis((17.4,17.9,0,5))
#ax1.set_yscale=("log")


ax2 = fig.add_subplot(222)
ax2.scatter(MCMC_raw[:,1], MCMC_raw[:,4], s=0.1)
ax2.set_xlabel("$\log_{10}(M_0)$",fontsize=20)
#ax2.set_ylabel(" $\chi^2$ ",fontsize=20)
#ax2.set_ylim(0,5)
#ax2.axis((10.90,11.5,0,5))
#ax2.set_yscale=("log")
ax2.yaxis.set_major_formatter(plt.NullFormatter())

ax3 = fig.add_subplot(223)
ax3.scatter(MCMC_raw[:,2], MCMC_raw[:,4], s=0.1)
ax3.set_xlabel(r" $\beta $  ",fontsize=20)
ax3.set_ylabel(" $\chi^2$ ",fontsize=20)
#ax3.axis((0.0,2.0,0,5))
#ax3.set_ylim(0,5)
#ax3.set_yscale=("log")
#ax.yaxis.set_major_formatter(plt.NullFormatter())

ax4 = fig.add_subplot(224)
ax4.scatter(MCMC_raw[:,3], MCMC_raw[:,4], s=0.1)
ax4.set_xlabel(r"$\gamma$",fontsize=20)
#ax4.set_ylabel(" $\chi^2$ ",fontsize=20)
ax4.yaxis.set_major_formatter(plt.NullFormatter())




################## Histograms #############################

fig = figure(figsize=(10, 5))
fig.subplots_adjust(hspace=0.20, wspace=0.0,top=1.5, right=1.0, bottom=0.0)

ax1 = fig.add_subplot(221)
ax1.hist(MCMC_raw[:,0],bins=30)

ax2 = fig.add_subplot(222)
ax2.hist(MCMC_raw[:,1],bins=30)

ax3 = fig.add_subplot(223)
ax3.hist(MCMC_raw[:,2],bins=30)

ax4 = fig.add_subplot(224)
ax4.hist(MCMC_raw[:,3],bins=30)

ax1.axvline(x=L_min);     ax1.axvline(x=average(MCMC_raw[:,0]), color="red")
ax1.axvline(x=percentile(MCMC_raw[:,0], q=25), color="green")
ax1.axvline(x=percentile(MCMC_raw[:,0], q=75), color="green")

ax2.axvline(x=M_min);     ax2.axvline(x=average(MCMC_raw[:,1]), color="red")
ax2.axvline(x=percentile(MCMC_raw[:,1], q=25), color="green")
ax2.axvline(x=percentile(MCMC_raw[:,1], q=75), color="green")
ax3.axvline(x=beta_min);  ax3.axvline(x=average(MCMC_raw[:,2]), color="red")
ax3.axvline(x=percentile(MCMC_raw[:,2], q=25), color="green")
ax3.axvline(x=percentile(MCMC_raw[:,2], q=75), color="green")
ax4.axvline(x=gamma_min); ax4.axvline(x=average(MCMC_raw[:,3]), color="red")
ax4.axvline(x=percentile(MCMC_raw[:,3], q=25), color="green")
ax4.axvline(x=percentile(MCMC_raw[:,3], q=75), color="green")

ax1.set_xlabel("$\log_{10}(L_0)$",fontsize=20)
ax2.set_xlabel("$\log_{10}(M_0)$",fontsize=20)
ax3.set_xlabel(r" $\beta $  ",fontsize=20)
ax4.set_xlabel(r"$\gamma$",fontsize=20)
```
