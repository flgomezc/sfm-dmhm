```python
A = range(0,64)
#output = [open( '../results/%i.dat' %filenumber, 'w') for filenumber in A]
BestPar  = open( '../analysis/best_parameters.dat', 'w')

for DATA in A:

    STR = '../results/'+str(DATA)+'.dat'
    MCMC_raw = np.loadtxt(STR,usecols=(0,1,2,3,4,5), skiprows=0); 
    
    BoxLength = 250.0 # Mpc h^{-1}

    ChiSqr_min= MCMC_raw[:,4].min()
    index = where( MCMC_raw[:,4] == MCMC_raw[:,4].min() )[0] 
    L_min     = MCMC_raw[index,0][0]
    M_min     = MCMC_raw[index,1][0]
    beta_min  = MCMC_raw[index,2][0]
    gamma_min = MCMC_raw[index,3][0]
    ChiSqr_min= MCMC_raw[index,4][0]
    COLS_min  = MCMC_raw[index,5][0]

    BestPar.write( str(L_min)     +'\t'+
               str(M_min)     +'\t'+
               str(beta_min)  +'\t'+
               str(gamma_min) +'\t'+
               str(ChiSqr_min)+'\t'+
               str(COLS_min)  +'\t'+
               str(DATA)  +'\t')
    
    for i in range(4):
        for j in range(3):
            Q = 25*(j+1)
            BestPar.write(str(percentile(MCMC_raw[:,i],q=Q))+'\t')
    BestPar.write('\n')
    
#    STR2 = '../analysis/'+str(DATA)+'.dat'
#    output = open( STR2, 'w' )
#    for i in range(MCMC_raw.shape[0]):
#        if ( (MCMC_raw[i,0] <= L_min + 0.1)&(MCMC_raw[i,0] >= L_min - 0.1)):
#            if ( (MCMC_raw[i,1] <= M_min + 0.05)&(MCMC_raw[i,1] >= M_min - 0.05)):
#                if ( (MCMC_raw[i,2] <= beta_min + 0.1)&(MCMC_raw[i,2] >= beta_min - 0.1)):
#                    if ( (MCMC_raw[i,3] <= gamma_min + 0.1)&(MCMC_raw[i,3] >= gamma_min - 0.1)):
#                        output.write(str(MCMC_raw[i,0])+"\t"+
#                                     str(MCMC_raw[i,1])+"\t"+
#                                     str(MCMC_raw[i,2])+"\t"+
#                                     str(MCMC_raw[i,3])+"\t"+
#                                     str(MCMC_raw[i,4])+"\t"+
#                                     str(MCMC_raw[i,5])+"\n")
#    output.close()

BestPar.close()
```
