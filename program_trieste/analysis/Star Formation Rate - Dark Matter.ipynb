```python
%pylab inline

def Luminosity(M,L_0,M_0,beta,gamma):
    return L_0*M*( (M/M_0)**(-beta) + (M/M_0)**gamma)**(-1)

def StarFormationRate(M,L_0,M_0,beta,gamma):
    return Luminosity(M,L_0,M_0,beta,gamma) * 1.4e-28
```

```python
Param_Scattering = np.loadtxt('../analysis/best_parameters.dat',usecols=(0,1,2,3), skiprows=0); 

L_0 = Param_Scattering[:,0]
M_0 = Param_Scattering[:,1]
beta = Param_Scattering[:,2]
gamma = Param_Scattering[:,3]
```

```python
#############################################
#                                           #
#           Plotting SFR function           #
#                                           #
#############################################
Mass = linspace(10,14,50)

figure(figsize(9,7))
for k in range(L_0.size):
    SFR = zeros(Mass.size)
    for i in range(Mass.size):
        SFR[i] = StarFormationRate(10**Mass[i],10**L_0[k],10**M_0[k],beta[k],gamma[k])
    loglog(10**Mass,SFR)
xlabel("DMH mass$[M_\odot]$", fontsize=20)
ylabel("SFR $[M_\odot yr^{-1}]$", fontsize=20)

savefig("star_formation_rate.pdf")
```
